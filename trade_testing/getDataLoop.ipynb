```python
import csv, time, requests, json, datetime

timeStampArray = []
usdPriceArray = []
marketCapArray = []
volumeArray = []
hourChangeArray = []
dayChangeArray = []
weekChangeArray = []
iterations = 0
maxIterations = 240
iterationGap = 60

coinName = input("Enter name of coin: ")
coinname = coinName.lower()
coinname = coinname.replace(" ", "-")

cmcTick = requests.get('https://api.coinmarketcap.com/v1/ticker/?limit=1000')
cmcText = cmcTick.text
cmcPrice = json.loads(cmcText)

def getTimeStamp():
    ts = time.time()
    st = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
    return st

def cmcPriceName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["price_usd"]
        i = i + 1
    return "INVALID COIN"

def cmcMCName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["market_cap_usd"]
        i = i + 1
    return "INVALID COIN"

def cmcVolumeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["24h_volume_usd"]
        i = i + 1
    return "INVALID COIN"

def cmcHourChangeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["percent_change_1h"]
        i = i + 1
    return "INVALID COIN"

def cmcDayChangeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["percent_change_24h"]
        i = i + 1
    return "INVALID COIN"

def cmcWeekChangeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["percent_change_7d"]
        i = i + 1
    return "INVALID COIN"
    

while (iterations < maxIterations):
    cmcTick = requests.get('https://api.coinmarketcap.com/v1/ticker/?limit=1000')
    cmcText = cmcTick.text
    cmcPrice = json.loads(cmcText)
    with open(coinName.lower() + 'PriceData.csv','r') as csvfile:
        priceReader = csv.reader(csvfile, delimiter=',')
        i = 0
        for row in priceReader:
            if i == 0:
                timeStampArray = row
            elif i == 1:
                usdPriceArray = row
            elif i == 2:
                marketCapArray = row
            elif i == 3:
                volumeArray = row
            elif i == 4:
                hourChangeArray = row
            elif i == 5:
                dayChangeArray = row
            elif i == 6:
                weekChangeArray = row
            i = i + 1

    timeStampArray.append(getTimeStamp())
    usdPriceArray.append(cmcPriceName(coinname))
    marketCapArray.append(cmcMCName(coinname))
    volumeArray.append(cmcVolumeName(coinname))
    hourChangeArray.append(cmcHourChangeName(coinname))
    dayChangeArray.append(cmcDayChangeName(coinname))
    weekChangeArray.append(cmcWeekChangeName(coinname))
        

    with open(coinName.lower() + 'PriceData.csv', 'w', newline='') as csvfile:
        priceWriter = csv.writer(csvfile, delimiter=',')
        priceWriter.writerow(timeStampArray)
        priceWriter.writerow(usdPriceArray)
        priceWriter.writerow(marketCapArray)
        priceWriter.writerow(volumeArray)
        priceWriter.writerow(hourChangeArray)
        priceWriter.writerow(dayChangeArray)
        priceWriter.writerow(weekChangeArray)
        
    time.sleep(iterationGap)
    iterations = iterations + 1

print("Done")
```
