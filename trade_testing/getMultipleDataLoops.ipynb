```python
#Tommy Bennett


#This program allows for price data to be scraped and stored within a csv file for later use.

#I used the csv library to read/write to csvfiles, the time and datetime library to accurately timestamp the price data,
#the requests library to get the json response from the pricing website (https://coinmarketcap.com/),
#the json library to extract the data from the json response.

import csv, time, requests, json, datetime

#Declaring the neccessary arrays
coinNameArray = []
timeStampArray = []
usdPriceArray = []
marketCapArray = []
volumeArray = []
hourChangeArray = []
dayChangeArray = []
weekChangeArray = []

#Declaring and initilizing variables
iterations = 0
maxIterations = 2016        #Number of times the data scraping loop will run
iterationGap = 60 * 5      #Number of seconds the loop will wait until it scapes data again
userInput = ""
numberOfCoins = 0

#Initializing array with the names of every coin for which data will be collected by the program
coinNameArray = ["0x","aelf","aeron","aion","airswap","appcoins","ark","bancor","basic-attention-token","binance-coin","bitcoin-cash","bitcoin","bitshares","bread","cardano","centra","chainlink","cindicator","coindash","cybermiles","dash","decentraland","digixdao","district0x","eidoo","enigma","enjin-coin","eos","ethereum-classic","ethereum","etherparty","ethlend","everex","funfair","gas","genesis-vision","gifto","gxshares","hshare","iconomi","icon","iota","komodo","kyber-network","lisk","litecoin","loopring","lunyr","metal","modum","monaco","monero","monetha","nav-coin","neblio","neo","nuls","oax","omisego","populous","power-ledger","qtum","quantstamp","raiden-network-token","request-network","ripio-credit-network","ripple","salt","simple-token","sonm","status","stellar","storj","stratis","substratum","tierion","time-new-bank","triggers","tron","vechain","verge","vibe","viberate","wabi","walton","waves","wings","zcash","zcoin"]

#Allows user to input the names of the coins they would like data collected for manually
#while (userInput != "stop"):
    #userInput = input("Enter name of coin (enter stop when done): ")
    #coinname = userInput.lower()
    #coinname = coinname.replace(" ", "-")
    #if (coinname != "stop"):
        #coinNameArray.append(coinname)

#Waypoint confirmation
print("Coins Loaded...\n")

#Calls to the website, receives a json response, and turns the json response into a Python object
cmcTick = requests.get('https://api.coinmarketcap.com/v1/ticker/?limit=1000')
cmcText = cmcTick.text
cmcPrice = json.loads(cmcText)

#Returns current time(Year-Month-Day Hour-Minute-Time)
def getTimeStamp():
    ts = time.time()
    st = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
    return st

#Returns the price of the coin that corresponds with the name passed through
def cmcPriceName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["price_usd"]
        i = i + 1
    return "INVALID COIN"

#Returns the marketcap of the coin that corresponds with the name passed through
def cmcMCName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["market_cap_usd"]
        i = i + 1
    return "INVALID COIN"

#Returns the 24 hour volume of the coin that corresponds with the name passed through
def cmcVolumeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["24h_volume_usd"]
        i = i + 1
    return "INVALID COIN"

#Returns the past hour's change in price of the coin that corresponds with the name passed through
def cmcHourChangeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["percent_change_1h"]
        i = i + 1
    return "INVALID COIN"

#Returns the past day's change in price of the coin that corresponds with the name passed through
def cmcDayChangeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["percent_change_24h"]
        i = i + 1
    return "INVALID COIN"

#Returns the past week's change in price of the coin that corresponds with the name passed through
def cmcWeekChangeName(nameInput):
    i = 0
    while i < 1000:
        if (cmcPrice[i]["id"] == nameInput):
            return cmcPrice[i]["percent_change_7d"]
        i = i + 1
    return "INVALID COIN"
    
#Main loop that runs through the coinNameArray and scraps data for each coin within that array and stores it into a csv file.
#This loop continues until it reaches the maximum iteration value which is initialized at with the other variables
while (iterations < maxIterations):
    cmcTick = requests.get('https://api.coinmarketcap.com/v1/ticker/?limit=1000')
    cmcText = cmcTick.text
    cmcPrice = json.loads(cmcText)
        
    for name in coinNameArray:
        timeStampArray = []
        usdPriceArray = []
        marketCapArray = []
        volumeArray = []
        hourChangeArray = []
        dayChangeArray = []
        weekChangeArray = []
        with open(name.lower() + 'PriceData.csv','r') as csvfile:
            priceReader = csv.reader(csvfile, delimiter=',')
            i = 0
            for row in priceReader:
                if i == 0:
                    timeStampArray = row
                elif i == 1:
                    usdPriceArray = row
                elif i == 2:
                    marketCapArray = row
                elif i == 3:
                    volumeArray = row
                elif i == 4:
                    hourChangeArray = row
                elif i == 5:
                    dayChangeArray = row
                elif i == 6:
                    weekChangeArray = row
                i = i + 1

        timeStampArray.append(getTimeStamp())
        usdPriceArray.append(cmcPriceName(name))
        marketCapArray.append(cmcMCName(name))
        volumeArray.append(cmcVolumeName(name))
        hourChangeArray.append(cmcHourChangeName(name))
        dayChangeArray.append(cmcDayChangeName(name))
        weekChangeArray.append(cmcWeekChangeName(name))
        

        with open(name.lower() + 'PriceData.csv', 'w', newline='') as csvfile:
            priceWriter = csv.writer(csvfile, delimiter=',')
            priceWriter.writerow(timeStampArray)
            priceWriter.writerow(usdPriceArray)
            priceWriter.writerow(marketCapArray)
            priceWriter.writerow(volumeArray)
            priceWriter.writerow(hourChangeArray)
            priceWriter.writerow(dayChangeArray)
            priceWriter.writerow(weekChangeArray)
        
    #Waypoint confirmation
    print("Iteration" + iteration + "complete...\n")
    
    time.sleep(iterationGap)
    iterations = iterations + 1

print("Done")
```
