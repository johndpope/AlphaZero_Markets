```python
import csv
import time
import datetime

coinNameArray = []
timeInFuture = 1
changeTime0 = datetime.timedelta(hours=timeInFuture)
changeTime1 = datetime.timedelta(hours=timeInFuture,minutes=5)


coinNameArray = ["0x","aelf","aeron","aion","airswap","ark","bancor","basic-attention-token","binance-coin","bitcoin-cash","bitcoin","bitshares","bread","cardano","centra","chainlink","cindicator","coindash","cybermiles","dash","decentraland","digixdao","district0x","eidoo","enigma","enjin-coin","eos","ethereum-classic","ethereum","etherparty","ethlend","everex","funfair","gas","genesis-vision","gifto","gxshares","hshare","iconomi","icon","iota","komodo","kyber-network","lisk","litecoin","loopring","lunyr","metal","modum","monaco","monero","monetha","nav-coin","neblio","neo","nuls","oax","omisego","populous","power-ledger","qtum","quantstamp","raiden-network-token","request-network","ripio-credit-network","ripple","salt","simple-token","sonm","status","stellar","storj","stratis","substratum","tierion","time-new-bank","triggers","tron","vechain","verge","vibe","viberate","wabi","walton","waves","wings","zcash","zcoin"]

for name in coinNameArray:
    
    #Declaring the neccessary arrays
    tempTimeStampArray = []
    tempUsdPriceArray = []
    tempMarketCapArray = []
    tempVolumeArray = []
    tempHourChangeArray = []
    tempDayChangeArray = []
    tempWeekChangeArray = []
    tempNextChangeArray = []
    coinNameArray = []
    timeStampArray = []
    usdPriceArray = []
    marketCapArray = []
    volumeArray = []
    hourChangeArray = []
    dayChangeArray = []
    weekChangeArray = []
    nextChangeArray = []
    
    with open(name.lower() + 'PriceData.csv','r') as csvfile:
        priceReader = csv.reader(csvfile, delimiter=',')
        i = 0
        for row in priceReader:
            if i == 0:
                timeStampArray = row
            elif i == 1:
                usdPriceArray = row
            elif i == 2:
                marketCapArray = row
            elif i == 3:
                volumeArray = row
            elif i == 4:
                hourChangeArray = row
            elif i == 5:
                dayChangeArray = row
            elif i == 6:
                weekChangeArray = row
            i = i + 1
        
    tempTimeStampArray.append("Time Stamp")
    tempUsdPriceArray.append("Price")
    tempMarketCapArray.append("Market Cap")
    tempVolumeArray.append("Volume")
    tempHourChangeArray.append("Hour Change")
    tempDayChangeArray.append("Day Change")
    tempWeekChangeArray.append("Week Change")
    tempNextChangeArray.append("Future Change")
    
    _count = 0
    _count1 = 0
    
    for time in timeStampArray:
        time_object = datetime.datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
        for time2 in timeStampArray:
            time2_object = datetime.datetime.strptime(time2, '%Y-%m-%d %H:%M:%S')
            deltaTimeObject = time2_object - time_object
            secondsPast = deltaTimeObject.total_seconds()
            if (secondsPast >= changeTime0.total_seconds() and secondsPast < changeTime1.total_seconds()):
                tempTimeStampArray.append(timeStampArray[_count])
                tempUsdPriceArray.append(usdPriceArray[_count])
                tempMarketCapArray.append(marketCapArray[_count])
                tempVolumeArray.append(volumeArray[_count])
                tempHourChangeArray.append(hourChangeArray[_count])
                tempDayChangeArray.append(dayChangeArray[_count])
                tempWeekChangeArray.append(weekChangeArray[_count])
                tempNextChangeArray.append(hourChangeArray[_count1])
            _count1 = _count1 + 1
        _count1 = 0
        _count = _count + 1
        
    with open(name.lower() + 'PriceData.csv', 'w', newline='') as csvfile:
        priceWriter = csv.writer(csvfile, delimiter=',')
        priceWriter.writerow(tempTimeStampArray)
        priceWriter.writerow(tempUsdPriceArray)
        priceWriter.writerow(tempMarketCapArray)
        priceWriter.writerow(tempVolumeArray)
        priceWriter.writerow(tempHourChangeArray)
        priceWriter.writerow(tempDayChangeArray)
        priceWriter.writerow(tempWeekChangeArray)
        priceWriter.writerow(tempNextChangeArray)
    
    print(name + " data cleaned...")

print("Done")
    
```
